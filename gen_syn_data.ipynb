```python
import matplotlib.pyplot as plt
import networkx as nx
import random

curr_node_num = 1

open("datasets/graph/graph_A.txt", "w").close()

open("datasets/graph/graph_graph_indicator.txt", "w").close()
open("datasets/graph/graph_graph_labels.txt", "w").close()
graph_indicator = 1
random_num = random.randint(5, 30)
random_graphs = []
for _ in range(250):
    n = random_num  # 10 nodes
    m = random_num  # 20 edges
    # seed = 42  # seed random number generators for reproducibility

    # Use seed for reproducibility
    G = nx.gnm_random_graph(n, m)
    random_graphs.append(G)
    # some properties
    # print("node degree clustering")
    # for v in nx.nodes(G):
    #     print(f"{v} {nx.degree(G, v)} {nx.clustering(G, v)}")

    # print()
    print(nx.generate_adjlist(G))
    pairs = []

    for line in nx.generate_adjlist(G):
        # Split the line into nodes
        nodes = line.split()
        # The first node is the source, and the rest are target
        source = nodes[0]
        targets = nodes[1:]
        
        # Generate (source, target) pairs
        for target in targets:
            print(str(int(source) + curr_node_num))
            pairs.append((str(int(source) + curr_node_num), str(int(target) + curr_node_num)))
    print(pairs)
    # Write the result to a file
    with open("datasets/graph/graph_A.txt", "a") as f:
        for pair in pairs:
            f.write(f"{pair[0]}, {pair[1]}\n")

    with open("datasets/graph/graph_graph_indicator.txt", "a") as f:
        for pair in pairs:
            f.write(f"{graph_indicator}\n")
    with open("datasets/graph/graph_graph_labels.txt", "a") as f:
        f.write(f"0\n")
        graph_indicator += 1
    curr_node_num += n
```
